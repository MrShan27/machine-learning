```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('ev_charging_patterns.csv')
```

```python
data
```

```python
data=data.drop(columns=["User ID","Charging Station ID","Charging Start Time","Charging End Time","Time of Day","Day of Week","Temperature (Â°C)","Energy Consumed (kWh)","Charging Rate (kW)","Distance Driven (since last charge) (km)"])
```

```python
data
```

```python
data['Vehicle_Model']=data['Vehicle_Model'].replace({"BMW i3":0,"Hyundai Kona":1,"Chevy Bolt":2,"Nissan Leaf":3,"Tesla Model 3":4})
data['Charging_Station_Location']=data['Charging_Station_Location'].replace({"Houston":0,"San Francisco":1,"Los Angeles":2,"New York":3,"Chicago":4})
data['Charger_Type']=data['Charger_Type'].replace({"DC Fast Charger":0,"Level 1":1,"Level 2":2})
data['User_Type']=data['User_Type'].replace({"Commuter":0,"Casual Driver":1,"Long-Distance Traveler":2})
```

```python
data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
independent=data[["Vehicle_Model","Battery_Capacity_kWh","Charging_Station_Location","Charging_Duration_hours","Charging_Cost_USD","State_of_Charge_Start_%","State_of_Charge_End_%","Charger_Type","User_Type"]]
dependent=data[["Vehicle_Age_years"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVR
Regressor = SVR(kernel='rbf') 
Regressor.fit(X_train, Y_train)
```

```python
Y_pred=Regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
print("r_score=",r_score)
```
